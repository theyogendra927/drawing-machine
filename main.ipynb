```python
import cv2
import numpy as np
```

```python
x=700
y=700
screen=np.zeros((x,y,3)).astype(np.uint8)
```

```python
r1=10
r2=10
for theta in range (0,180):
    cy1=r1*np.cos(np.pi/180*theta)
    cx1=r1*np.sin(np.pi/180*theta)
    for phy in range (0,180):
        cy2=cy1+ r2*np.cos(np.pi/180*phy)
        cx2=cx1+ r2*np.cos(np.pi/180*phy)
        cv2.circle(screen,(cx2,cy2),2,(0,255,0),2)
```
