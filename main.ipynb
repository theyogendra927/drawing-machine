```python
import cv2
import numpy as np
import time
```

```python
x=800
y=800
origin=(x//2,y//2)
```

```python
def calculator(x,y,r1,r2,origin):
    x=x-origin[0]
    y=y-origin[1]
    theta_diff=np.arccos((x*x+y*y-r1*r1-r2*r2)/(2*r1*r2))#the eqn should result that x and y satisfy +-sign acc. to new origin
    divider=np.sqrt((r2+r1*np.cos(theta_diff))**2+(r1*np.sin(theta_diff))**2)
    theta2=np.arcsin(y/divider)-np.arccos((r2+r1*np.cos(theta_diff))/divider)
    theta=theta_diff+theta2
    phy=theta2-theta    
    return (theta,phy) 
```

```python
r1=100
r2=100
screen=np.ones((x,y)).astype(np.uint8)*255
for theta in range(180,360):
    cx1=r1*np.cos(np.pi/180*theta)+origin[1]
    cy1=r1*np.sin(np.pi/180*theta)+origin[0]
    for phy in range (0,180):
        cx2=cx1+ r2*np.cos(np.pi/180*(phy+theta))
        cy2=cy1+ r2*np.sin(np.pi/180*(phy+theta))
        print(int(cx2),int(cy2))
        theta,phy=calculator(x,y,r1,r2,origin)
        cx2=r1*np.cos(theta)+origin[1]+ r2*np.cos((phy+theta))
        cy2=r1*np.sin(theta)+origin[0]+ r2*np.sin((phy+theta))
        print(cx2,cx1)
        cv2.circle(screen,(int(cx2),int(cy2)),2,(0,255,0),2)
        cv2.imshow("logic",screen)
    cv2.waitKey(1)
```

```python
while True:
    x=int(input("enter x"))
    y=int(input("enter y"))
    theta,phy=calculator(x,y,r1,r2,origin)
    cx2=r1*np.cos(theta)+origin[1]+ r2*np.cos((phy+theta))
    cy2=r1*np.sin(theta)+origin[0]+ r2*np.sin((phy+theta))
    print(cx2,cx1)
```
